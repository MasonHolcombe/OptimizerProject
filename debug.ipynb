```python
from Network import Network
from Layer import Dense, ReLU, Sigmoid, Linear, Softmax
import numpy as np
```

```python
input = np.array([0,1,2,3,4,5])

NN = Network(input)
NN.add(Dense(len(input), 0.001), activation=ReLU())
NN.add(Dense(1, 0.001), activation=Sigmoid())
# NN.forward()
NN.compile()
```

```python
temp = []
for _ in range(100):
    NN = Network(input)
    NN.add(Layer(len(input), ReLU()))
    NN.add(Layer(5, ReLU()))
    NN.add(Layer(6, ReLU()))
    NN.add(Layer(1, Sigmoid()))
    NN.add(Layer(1, Linear()))
    temp.append(NN.forward()[0])
temp
```

```python
from Network import Network
from Layer import Dense, ReLU, Sigmoid, Linear, Softmax, Input
from Loss import Categorical_CE
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv("IRIS.csv")
X = iris[['sepal_length','sepal_width','petal_length','petal_width']]
y = iris['species']
y = y.map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
print(f'X Shape: {X.shape}\nY Shape: {np.array(y).shape}')
```

```python
# input = np.array([0,1,2,3,4,5])
# size = input.shape[0]
# input = np.reshape(input, (input.shape[0], 1))
y = np.array(y)
X = np.array(X)

NN = Network(X.T, Categorical_CE())
NN.add(Input(4, 0.001), activation=ReLU())
NN.add(Dense(5, 0.001), activation=ReLU())
NN.add(Dense(3, 0.001), activation=Softmax())

NN.compile(debug=0)
#NN.train(X=input, y=np.loss=Categorical_CE)
print(NN)
pred = NN.forward()
print(f'Network Loss: {NN.calcLoss(y, pred)}')
```

```python
def one_hot_encode(vector):
    unique_values = list(set(vector))
    unique_values.sort()
    
    encoded_vectors = []
    
    for value in vector:
        one_hot_vector = [0] * len(unique_values)
        index = unique_values.index(value)
        one_hot_vector[index] = 1
        encoded_vectors.append(one_hot_vector)
    
    return encoded_vectors
```
